Demonstration of a Boxplot of the error on the NARMA task as a function of the reservoir size (default from $N = 10$ to $N = 18$). The number of samples taken per box is 10, it can be increased at the expend of higher computational time. With the default number of samples the simulations should take around 30 seconds to complete.

```python
import numpy as np
import matplotlib.pyplot as plt

#Set seed
np.random.seed(0)

#PARAMETERS
# T: beam-splitter transmissivity
# d: NARMA delay
T = 0.4
d = 5
#Number of samples
samples = 10
#Reservoir sizes
N_vec = [10,12,14,16,18]

box_data = []
for i in range(len(N_vec)):
    error_arr = np.zeros(samples,float)
    N = N_vec[i]
    for j in range(samples):
        error_arr[j] = ut.get_NARMA(N,T,d)
    box_data.append(error_arr)

plt.boxplot(box_data)
plt.xticks([1,2,3,4,5],N_vec,fontsize=15)
plt.xlabel('Reservoir Size',fontsize=15)
plt.ylabel('NMSE (NARMA%i)' %(d),fontsize=15)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
